# RIDGE REGRESSION - LASSO MODEL

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
X_train_s = scaler.transform(X_train)
```

```python
X_test_s = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha = 0.5)#lambda is 0.5
lm_r.fit(X_train_s,y_train)
```

```python
r2_score(y_test,lm_r.predict(X_test_s))
```

```python
from sklearn.model_selection import validation_curve
```

```python
validation_curve?
#estimater  is ridge or lasso...param_name is aplha...
#param_range is multiple values of alpha...scoring is r2....
```

```python
param_range = np.logspace(-2,8,100)#100 values b/w 10^-2 to 10^8
```

```python
param_range
```

```python
train_scores , test_scores = validation_curve( estimator=Ridge(),
    X=X_train_s,
    y=y_train,
    param_name="alpha",
    param_range=param_range,
    scoring='r2')
```

```python
print(train_scores)
print( test_scores)#vc is running k fold validation behind the scenes
```

```python
train_mean = np.mean(train_scores , axis = 1)
```

```python
test_mean = np.mean(test_scores , axis = 1)
```

```python
train_mean
```

```python
max(train_mean)
```
